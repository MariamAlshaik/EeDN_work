```python
import networkx as nx
graph_list = []
```

```python
G=nx.Graph()
G.add_node(0, label='M', entity='first')
G.add_node(1, label='A', entity='second')
G.add_node(2, label='R', entity='third')
G.add_node(3, label='Y', entity='fourth')
G.add_edge(0,1, label='f', weight='0')
G.add_edge(1,2, label='s', weight='1')
G.add_edge(2,3, label='t', weight='2')
#G.add_edge(3,4, label='f', weight='3')
graph_list += [G.copy()]

G=nx.Graph()
G.add_node(0, label='M', entity='first')
G.add_node(4, label='A', entity='beforLast')
G.add_node(5, label='M', entity='last')
G.add_node(6, label='FR', entity='mom')
G.add_edge(4,5, label='f', weight='4')
G.add_edge(5,0, label='s', weight='5')
G.add_edge(0,6, label='d6', nesting=True, weight='6')
G.add_edge(5,6, label='d6', nesting=True, weight='6')
graph_list += [G.copy()]
```

```python
from eden.util import display
for g in graph_list:
    display.draw_graph(g, size=20, node_size=800, node_border=4, layout='spectral', secondary_vertex_label='entity', secondary_edge_label='weight')
```

```python
%%time
from eden.graph import Vectorizer
vectorizer = Vectorizer(complexity=2, n=3)
vectorizer.fit(graph_list)
X = vectorizer.transform(graph_list)
y=[1]*3+[2]*3
print 'instances: %d \n features: %d with Avg of %d features per instance' %(X.shape[0], X.shape[1], X.getnnz()/X.shape[0])
```
