## Auxiliary functions
    - Generate rondome dataset with required length.
    - Swap the position of two charecter in the previosly generated dataset.
    - Generate graphs from the generated dataset.

```python
%matplotlib inline
```

```python
##genetate random dataset from a-->z
import random

#swap the position of two characters randomely
def swap_two_characters(string_l):
    items = list(string_l)
    rand1 = random.randint(0,len(string_l)-1)
    rand2 = random.randint(0,len(string_l)-1)
    items[rand1], items[rand2] = items[rand2], items[rand1]
    return ''.join(items)

mkstring = lambda(x): "".join(map(chr, (ord('a')+(y) for y in range(x))))

#generate random words
def words_generation(n_strings=10, n_characters=5, string_generator=None):  
    words = []
    word = string_generator(n_characters)
    for j in range(n_strings):
            word = swap_two_characters( word )
            words.append(word)
    return words
```

```python
import networkx as nx
#generate graphs from a word
def string_to_graph(word):
    G = nx.Graph()
    for i,letter in enumerate(word):
        G.add_node(i, label = letter)
        if i > 0:
            G.add_edge(i-1, i, label = 'x')
    return G
```

## Experiment

```python
#The generation of the dataset
words = words_generation(n_strings=10, n_characters=10, string_generator=mkstring)
print words
```

```python
#The display of the generated graphs
from eden.util import display
for word in words:
    word_graph = string_to_graph(word)
    display.draw_graph(word_graph)
```
